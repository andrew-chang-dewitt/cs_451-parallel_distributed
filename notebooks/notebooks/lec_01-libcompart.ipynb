```python
# Notebook to use libcompart on FABRIC
# Nik Sultana
# http://www.cs.iit.edu/~nsultana1/teaching/S26CS451/

from fabrictestbed_extensions.fablib.fablib import FablibManager as fablib_manager

fablib = fablib_manager(project_id="f5b4fc7d-978f-45be-9530-b38db8ef5046")

slice_name = 'libcompart'
```

```python
image = 'default_ubuntu_22'
site = "LOSA"
slice = fablib.new_slice(name = slice_name)
n1 = slice.add_node(name = "n1", image = image, cores = 2, ram = 4, disk = 9, site = site)
n1_iface1 = n1.add_component(model = "NIC_Basic", name = "iface1").get_interfaces()[0]
n2 = slice.add_node(name = "n2", image = image, cores = 2, ram = 4, disk = 9, site = site)
n2_iface1 = n2.add_component(model = "NIC_Basic", name = "iface1").get_interfaces()[0]
net = slice.add_l2network(name = "net", interfaces=[n1_iface1, n2_iface1], type = "L2Bridge")
slice.submit()
```

```python
slice = fablib.get_slice(slice_name)
```
